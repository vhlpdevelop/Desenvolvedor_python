```python
#Normalização e padronização de dados
#Exemplos de contexto e aplicação
#Algoritmos sensiveis a escala
#Comparação de atributos
#Analise Estatistica
#Redução de dimensionalidade(PCA)
#Algoritmos que utilizam medidas de distancia(Como k-means, k-NN e algoritmos de clustering)
#Todos acima de beneficiam da padronização de dados

#Primeiro, MinMaxScaler(NORMALIZAÇÃO DE DADOS)
#É uma tecnica de pre-processamento, comum usada em analise de dados e no treinamento de modelos de aprendizado de máquina

#Ele calcula o valor minimo e maximo de cada variavel e em seguida aplica transformação linear aos dados
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#Dados de exemplo

dados = {
    'idade':[25,30,35,40,45],
    'salario':[30000,40000,50000,60000,70000],
    'pontuacao':[50,60,70,80,90]
}

df = pd.DataFrame(dados)

#Criando instancia do minmaxscaler
scaler = MinMaxScaler()

#aplicando minmaxscaler
dados_normalizados = scaler.fit_transform(df)
print('Dados normalizados pelo scaler')
print(dados_normalizados)
#Exibindo dataframe original

df_normalizado = pd.DataFrame(dados_normalizados, columns=df.columns)

print('\n DataFrame Original:\n',df)

print('\n DataFrame Normalizado: \n', df_normalizado) 
#Exibindo dataframe normalizado
```

```python
#Padronização de dados com StandardScaler
#Tecnica de pre-processamento, usada em analise e treinamento de aprendizado de maquina
#Padroniza variaveis, transformando-as para que tenham média zero e desvio padrão igual a um.
#Ele calcula a media e o desvio padrao de cada variavel, em seguida, aplica uma transformação de dados, subtraindo a media e dividindo pelo desvio padrao
from sklearn.preprocessing import StandardScaler
dados = {
    'idade':[25,30,35,40,45],
    'salario':[30000,40000,50000,60000,70000],
    'pontuacao':[50,60,70,80,90]
}
df = pd.DataFrame(dados)

#Criando instancia Scaler
stscaler = StandardScaler()
#Padronizando dataframe
dados_padronizados = stscaler.fit_transform(df)
#Exibindo dataframe original

df_padronizado = pd.DataFrame(dados_padronizados, columns=df.columns)

print('\n DataFrame Original:\n',df)

print('\n DataFrame padronizado: \n', df_padronizado) 
#Exibindo dataframe padronizado

#Padronização é diferente de normalização
```

```python
#RobustScaler, outro padronizador de dados
#Usado em Analise e treinamento de modelos de aprendizado de maquina. Padroniza variaveis tornando-as mais robustas a outliers e
#distribuições nao normais.
#Diferente do StandardScaler, ele utiliza estatisticas resistentes a outliers, como mediana e o intervalo interquartil(IQR) para padronizar
from sklearn.preprocessing import RobustScaler
dados = {
    'idade':[25,30,35,40,45,120],
    'salario':[30000,40000,50000,60000,70000,200000],
    'pontuacao':[50,60,70,80,90,30]
}
df = pd.DataFrame(dados)

#Criando instancia RobustScaler
rbscaler = RobustScaler()
#Padronizando dataframe
dados_padronizados = rbscaler.fit_transform(df)
#Exibindo dataframe original

df_padronizado = pd.DataFrame(dados_padronizados, columns=df.columns)

print('\n DataFrame Original:\n',df)

print('\n DataFrame padronizado: \n', df_padronizado) 
#Exibindo dataframe padronizado

#Padronização é diferente de normalização
```
