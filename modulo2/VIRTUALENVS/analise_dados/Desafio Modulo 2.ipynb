```python
#Usando dataframe carac


frequencia_cor_cabelo = df_clientes_carac['Cor do Cabelo'].value_counts()

# Identificando a cor de cabelo mais comum
cor_mais_comum = frequencia_cor_cabelo.idxmax()

print(f"A cor de cabelo mais comum entre os clientes é: {cor_mais_comum}")
df_clientes_carac.head()
```

```python
#Questao 3- Existe alguma correlação entre altura e o peso do cliente?
#df_clientes_carac.head()

#Usaremos pearson
# Calculando a correlação de Pearson
import scipy.stats as stats

import seaborn as sns

coef_corr, p_value = stats.pearsonr(df_clientes_carac['Altura (cm)'], df_clientes_carac['Peso (kg)'])
print(f"Coeficiente de correlação de Pearson: {coef_corr:.2f}")
print(f"valor de p_value:{p_value}")
# Interpretação
if coef_corr > 0:
    print("Há uma correlação positiva entre gastos em horas de estudo com as notas.")
elif coef_corr < 0:
    print("Há uma correlação negativa entre gastos em horas de estudo com as notas.")
else:
    print("Não há correlação linear entre gastos em horas de estudo com as notas.")
```

```python
#Questao 4- Qual é o estado com a maior média de salário entre os clientes?
media_salario_por_estado = df_clientes_info.groupby('Estado')['Salário'].mean()

# Identificando o estado com a maior média de salário
estado_maior_media_salario = media_salario_por_estado.idxmax()

maior_media_salario = media_salario_por_estado.max()

print(f"O estado com a maior média de salário é: {estado_maior_media_salario}")
print(f"Média de salário: {maior_media_salario}")
```

```python
#Questao 5
df_clone_info = df_clientes_info
#Antes será necessario tratar os dados em idade
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')

#Dataframe só de clientes e idade
df_client_idade = df_clone_info[['ID Cliente', 'Idade']]
#Aplicando imputer
#PRIMEIRO vem o treinamento
df_imputer = pd.DataFrame(imputer.fit_transform(df_client_idade), columns=df_client_idade.columns)
#Criado novo dataframe para visualizar
print('\n Conjunto de dados ausentes tratados')
print(df_imputer)

print('\n Dados ausentes em INFO')
print(df_imputer.isna().sum(), '\n')

import matplotlib.pyplot as plt
plt.hist(df_imputer['Idade'],
         bins=30,
         edgecolor='black')

#Personalizar
plt.xlabel('Idade')
plt.ylabel('Número de clientes')
plt.title('Distribuição da idade dos clientes')

plt.show()
```

```python
#Questao 7
df_clone_info = df_clientes_info
df_clone_info['Acima_5000'] = df_clone_info['Salário'] > 5000

# Calculando a proporção para cada estado
proporcoes = df_clone_info.groupby('Estado')['Acima_5000'].mean()

# Identificando o estado com maior proporção
estado_maior_proporcao = proporcoes.idxmax()

print(f"Estado com maior proporção de salários acima de 5000 reais: {estado_maior_proporcao}")
```

```python
#questao 9 Qual é a média de altura para as pessoas que residem no estado de MG, são solteiras e possuem filhos?
#info idcliente, estado, temfilho, estado civil
#carac idcliente, altura

#Clonar
df_clone_info = df_clientes_info
df_clone_carac = df_clientes_carac.drop_duplicates()

df_merged = pd.merge(df_clone_info, df_clone_carac, on='ID Cliente', how='inner')

#Criar filtro
filtro = (df_merged['Estado'] == 'MG') & \
         (df_merged['Estado Civil'] == 'Solteiro') & \
         (df_merged['Tem Filhos'] == 'Sim')

media_altura = df_merged.loc[filtro, 'Altura (cm)'].mean()

print(f"Média de altura para pessoas que residem em MG, são solteiras e possuem filhos: {media_altura:.2f} metros")
```
