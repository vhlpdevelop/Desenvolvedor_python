```python
#Codificação variáveis categóricas (OneHotEncoder)
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

dados = {
    'cor': ['vermelho', 'azul', 'verde', 'vermelho', 'verde'],
    'tamanho':['pequeno', 'medio', 'grande', 'medio', 'grande'],
    'formato':['quadrado', 'redondo', 'quadrado', 'redondo', 'quadrado']
}
df = pd.DataFrame(dados)

#Criando estancia
encoder = OneHotEncoder()

#fazer o fit
dados_codificados = encoder.fit_transform(df[['cor','tamanho', 'formato']]).toarray()

#Criando novo dataframe para visualização
df_codificado = pd.DataFrame(dados_codificados, columns=encoder.get_feature_names_out(['cor','tamanho','formato']))
print('DataFrame original')
print(df)
print('DataFrame codificado')
df_codificado
```

```python
#Utilizando OrdinalEncoder
#Tecnica de codificação para transformar variaveis categoricas em dados númericos, atribuindo valores ordinais as categorias
from sklearn.preprocessing import OrdinalEncoder

dados = {
    'Nome':['Leandro','Daniele','Gilberto','Roberta','Joao'],
    'Idade':[32,27,54,39,68],
    'rating_satisfacao':['Satisfeito', 'Muito Satisfeito', 'Muito Insatisfeito', 'Insatisfeito', 'Neutro']
}

df = pd.DataFrame(dados)
df

#Criando ordem de categoria
categorias = ['Muito Insatisfeito', 'Insatisfeito', 'Neutro', 'Satisfeito', 'Muito Satisfeito' ]
#Criando instancia do ordinalencoder
encoder = OrdinalEncoder(categories = [categorias])

#Codificando

df['rating_satisfacao_enconder'] = encoder.fit_transform(df[['rating_satisfacao']])

#Exibindo DataFrame modificado
df.head()

#Por fim ele praticamento adicionou a coluna rating para melhorar visualização e inseriu a nota do rating de acordo com a categoria
```

```python
#Utilizando labelEncoder
#Util quando temos variaveis categoricas com diferentes categorias e precisamos converte-las em valores numericos
#para que os algoritmos de aprendizado de máquina possam processá-las. Ele atribui um número unico a cada categoria
from sklearn.preprocessing import LabelEncoder

dados= {
    'tamanho':['pequeno','medio','grande','medio','pequeno'],
    'intensidade':['fraco','medio','forte','forte','fraco'],
    'classificacao':['baixo','medio','alto','medio','baixo']
}
df = pd.DataFrame(dados)

#Exibindo dataframe original
print('DataFrame Original')
print(df)

#Criando estancia labelencoder
encoder = LabelEncoder()


#Aplicando labelencoder em cada coluna usando repetição
for coluna in df.columns:
    df[coluna] = encoder.fit_transform(df[coluna])

#Exibindo dataframe codificado
print('DataFrame Codificado')
df.head()

```
