```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Dados de exemplos
x = [ 1,2,3,4,5 ]
y = [ 2,4,6,8,10 ]
```

```python
#Grafico de linha
#Usado para fazer analise de tendencia temporal, comparação de dados, previsao e projecao, e por fim monitoramento de metrica
grafico = plt.plot(x,y)
grafico
```

```python
#Personalizar os eixos
#Novo exemplo
x = [ 1,2,3,4,5 ]
y = [ 2,4,6,8,10 ]
#criação do grafico de linha
plt.plot(x, y, label='Serie 1')

#Personalização do eixo
plt.xlabel('Eixo X')

plt.ylabel('Eixo Y')

plt.title('Titulo do grafico')

plt.legend() #Aqui traz o label definido em plt.plot
```

```python
#Adicao de multiplas linhas no grafico
#Novo exemplo
x = [ 1,2,3,4,5 ]
y1 = [ 2,4,6,8,10 ]
y2 = [ 1,3,5,7,9 ]
y3 = [ 3,6,9,12,15 ]

#Criação do grafico de linhas com multiplas linhas
plt.plot(x,y1, label='Serie 1')
plt.plot(x,y2, label='Serie 2')
plt.plot(x,y3, label='Serie 3')
#Apos feito para cada Y
plt.xlabel('Eixo X')

plt.ylabel('Eixo Y')

plt.title('Grafico de linhas multiplas')

plt.legend()

plt.show()
```

```python
#Uso de diferentes estilos de linha e marcadores
x = [ 1,2,3,4,5 ]
y1 = [ 2,4,6,8,10 ]
y2 = [ 1,3,5,7,9 ]

#Criando linha com estilo
plt.plot(x,y1, linestyle='--', marker='o',label='Serie 1')
plt.plot(x,y2, linestyle=':', marker='s',label='Serie 2')

plt.xlabel('Eixo X')

plt.ylabel('Eixo Y')

plt.title('Grafico de linhas estilosas')

plt.legend()

plt.show()
```

```python
#Exemplo pratico envolvendo 2 produtos eletronicos e dataframe
dados = {
    'Ano':[2019,2020,2021,2022,2023],
    'Eletronicos':[100,120,150,140,180],
    'Vestuario':[80,90,110,100,140]
}
df_vendas = pd.DataFrame(dados)
df_vendas
```

```python
#Definir eixo X e Y
x_anos = df_vendas['Ano']
y_vendas_eletronicos = df_vendas['Eletronicos']
y_vendas_vestuario = df_vendas['Vestuario']

#Criando grafico
plt.plot(x_anos,y_vendas_eletronicos, linestyle='--', marker='o',label='Eletronicos')
plt.plot(x_anos,y_vendas_vestuario, linestyle=':', marker='s',label='Vestuario')

plt.xlabel('Ano')

plt.ylabel('Vendas')

plt.title('Comparação de vendas de Eletronicos e Vestuarios')

plt.legend()

plt.show()
```

```python
#GRAFICOS DE DISPERSÃO
#PERMITE IDENTIFICAR PADROES, TENDENCIA, ANOMALIAS NOS DADOS

#dados exemplo
x = [1,2,3,4,5]
y = [2,4,6,8,10]

#GRAFICO PODE SER UTILIZADO
#ANALISE DE CORRELAÇÃO, IDENTIFICAÇÃO DE OUTLIERS, COMPARAÇÃO DE GRUPOS, PREVISÃO E MODELAGEM E ANALISE DE SERIES TEMPORAIS

plt.scatter(x,y)
plt.xlabel('Variavel X')
plt.ylabel('Variavel Y')
plt.title('Grafico dispersão')
plt.show()
```

```python
#Novo grafico com base a outros dados
altura = [165,170,175,168,172,180,160,155,162,175]
peso = [60,65,70,63,68,75,55,50,58,70]

#Criação do grafico com cores
plt.scatter(altura,peso, c='red', marker='^')
#Personalização
plt.xlabel('cm')
plt.ylabel('kg')
plt.title('Grafico Altura x Peso')



#Exibição
plt.show()
```

```python
#Exemplo pratico
dados = {
    'Altura': [165,170,175,168,172,180,160,155,162,175],
    'Peso': [60,65,70,63,68,75,55,50,58,70],
    'Idade': [25,30,35,28,32,40,22,20,24,34]
}

df = pd.DataFrame(dados)

#Criação do grafico 
plt.scatter(df['Altura'], df['Peso'], c=df['Idade'], marker='o', s=80, alpha=0.8) #s=80 seria tamanho do marker, alpha seria transparencia

#Adicionando barra de cor
cbar= plt.colorbar()

cbar.set_label('Idade')

plt.show()
```

```python
#GRAFICO HISTOGRAMA

#USADOS EM ANALISE DE DISTRUBUIÇÃO, IDENTIFICAÇÃO DE INTERVALOS CRITICOS, DETECCAO DE ASSIMETRIA, ANALISE DE TENDENCIA E COMPARAÇÃO DE GRUPOS

dados = [10,15,20,22,25,28,30,32,35,38,40,42,45,50,55,60,65,70]
df = pd.DataFrame(dados, columns=['Valores'])

#criação do grafico

plt.hist(df['Valores'],
         bins=5,
         edgecolor='black')

#Personalizar
plt.xlabel('Valores')
plt.ylabel('Frequencias')
plt.title('Distribuição dos dados')

plt.show()
```

```python
#Outro exemplo com personalização de intervalos

intervalos = [0,20,40,60,80]
bins = len(intervalos)-1

#Agora criação do grafico

plt.hist(df['Valores'],
         bins=bins,
         edgecolor='black')
#Personalizar
plt.xlabel('Valores')
plt.ylabel('Frequencias')
plt.title('Distribuição dos dados')

#Personalizando intervalo
plt.xticks(intervalos)

plt.show()
```

```python
#Usando histogramas com numpy

import numpy as np
#dados
dados1= np.random.normal(50,10,1000) #Distribuição 1
dados2= np.random.normal(70,5,1000) #Distribuição 2

#Criação do histograma empilhado
plt.hist([dados1, dados2], 
        bins=20,
         color=['red', 'green'],
         edgecolor='black',
         label=['Distribuição 1', 'Distribuição 2'],
         stacked=True #VAR QUE EMPILHA
        )
plt.xlabel('Valores')
plt.ylabel('Frequencias')
plt.legend()
plt.title('Histograma Empilhado')

plt.show()
```

```python
#Grafico de pizza

categorias = ['A','B','C','D']
valores = [30,25,20,15]

#Criação do grafico
plt.pie(valores, labels=categorias)

plt.title('Grafico Pizza - Distruibuição por categoria')
plt.legend()
plt.show()
```

```python
#Destacando fatias em grafico pizza

categorias = ['A','B','C','D']
valores = [30,25,20,15]
destaque = [0.1,0,0,0]

#Cores
cores=['blue','green','orange','red']


#Criação do grafico
patches, texts, autotexts = plt.pie(valores, labels=categorias, explode=destaque, colors=cores, autopct='%1.f%%') #Autopct faz apresentar formato númerico da pizza

plt.title('Grafico Pizza - Distruibuição por categoria')
plt.legend()
#Configurando cores e estilos dos numeros

for text in autotexts:
    text.set_color('white')
    text.set_fontweight('bold')

plt.show()
```

```python
#ALTERANDO PALETA DE CORES
import matplotlib.cm as cm #Color map
categorias = ['A','B','C','D']
valores = [30,25,20,15]

#Definindo paleta
cores_pastel = cm.get_cmap('Pastel1')

plt.pie(valores, labels=categorias, colors=cores_pastel(range(len(categorias))), autopct='%1.1f%%')

plt.title('Grafico Pizza')
plt.legend()
plt.show()
```

```python
#Grafico de area

anos=[2010,2011,2012,2013,2014,2015,2016]
variavel=[100,120,150,130,140,160,180]

#CRIAÇÃO DO GRAFICO
plt.fill_between(anos,variavel, color='blue', alpha=0.3)
plt.xlabel('Ano')
plt.ylabel('Variavel')
plt.title('Grafico de area')

plt.show()
```

```python
#Personalização de cores
anos=[2010,2011,2012,2013,2014,2015,2016]
var1=[10,20,30,25,35,40,50]
var2=[20,15,25,30,40,35,45]
var3=[5,10,15,20,25,30,35]

#Criar figura e os eixos
fig, ax = plt.subplots()

#plotar areas
ax.stackplot(anos,var1, var2, var3, labels=['Var 1','Var 2', 'Var 3'],
             colors=['#FF7F50', '#6A5ACD','#FFD700'], alpha=0.7)

#Personalizando
ax.set_title('Grafico de Area')
ax.set_xlabel('Anos')
ax.set_ylabel('Valores')
ax.legend(loc='upper left') #manda a localização da legenda

plt.show()
```

```python
import numpy as np
#Simulação com dados
data={
    'Data': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    'Vendas':np.random.randint(20,50,size=365),
    'Lucro':np.random.uniform(8,15,size=365)
}
df = pd.DataFrame(data)

cores = ['#FF7F50','#6A5ACD']

fig,ax= plt.subplots (figsize=(10,5))

df.plot(x='Data', y=['Vendas', 'Lucro'], kind='area',color=cores, alpha=0.5, ax=ax) #Plotando com dataframe
ax.set_title('Analise de series temporais')
ax.set_xlabel('Data')
ax.set_ylabel('Valores')

plt.show()
```

```python
#Grafico boxspot, usado para distribuições, analise comporativa, (saber valor minimo, maximo, outline)

dados1=np.random.normal(0,1,100)
dados2=np.random.normal(0,2,100)
dados3=np.random.normal(0,3,100)

#Lista dos dados
dados = [dados1,dados2,dados3]
#Criando grafico
plt.boxplot(dados)

#Configurando rotulos e titulos
plt.xticks([1,2,3], ['Dados 1', 'Dados 2', 'Dados 3'])
plt.title('Grafico boxplot')
plt.xlabel('Variaveis')
plt.ylabel('Valores')
plt.show()
```

```python
#Grafico boxsplot personalizado
dados1=np.random.normal(0,1,100)
dados2=np.random.normal(0,2,100)
dados3=np.random.normal(0,3,100)

#Lista dos dados
dados = [dados1,dados2,dados3]
#Criando grafico
plt.boxplot(dados, patch_artist=True) #patch_artist 

#Configurando rotulos e titulos
plt.xticks([1,2,3], ['Dados 1', 'Dados 2', 'Dados 3'])
plt.title('Grafico boxplot')
plt.xlabel('Variaveis')
plt.ylabel('Valores')
plt.show()
```

```python
#Exemplo usando boxsplot
idades_masc = np.random.normal(30,5,100)
idades_fem = np.random.normal(35,5,100)

df = pd.DataFrame({
    'Masculino': idades_masc,
    'Feminino':idades_fem
})

#Criando grafico
df.plot(kind='box', vert=False, patch_artist=True, widths=0.3)

#Configurando rotulos e titulos

plt.title('Grafico boxplot de idade')
plt.xlabel('Idade')
plt.ylabel('Genero')
plt.show()
```

```python
#Grafico de barras e linhas usado em analise de vendas por regiao, desempenho de projetos

#Dados
anos = [2018,2019,2020,2021,2022]
receita_total=[50000,60000,70000,90000,95000]
receita_produto1=[20000,22000,25000,28000,30000]
receita_produto2=[30000,38000,45000,62000,65000]

data = {
    'Ano': anos,
    'Receita Produto 1': receita_produto1,
    'Receita Produto 2': receita_produto2,
    'Receita Total': receita_total
}
df = pd.DataFrame(data)

#Criando figura e eixo
fig, ax1=plt.subplots(figsize=(15,6))

#Plotando barras
df[ ['Ano', 'Receita Produto 1', 'Receita Produto 2']].plot(x='Ano', kind='bar', stacked='True', ax=ax1, color=['blue','green'], alpha=0.5)

#Configuração dos eixos

ax1.set_xlabel('Ano')
ax1.set_ylabel('Valores Receita')


#Criando eixo secundario
ax2=ax1.twinx()

#Plotando grafico de linha

df['Receita Total'].plot(kind='line', marker='o', color='red', ax=ax1, legend=True)
plt.plot(anos, receita_total, color='red', marker='o', label='Receita Total') #esse plot coloca as info ao lado direito do grafico da receita_total
#Configurando eixo secundario
ax2.set_ylabel('Receita Total')
plt.title('Receita Total por produto e ao longo dos anos')

#Ajustando layout

plt.tight_layout()

plt.show()
```
