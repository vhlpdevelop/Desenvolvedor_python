```python
import pandas as pd
df = pd.read_csv("risco_credito.csv")
```

```python
df.tail()
```

```python
df['risco'].unique()
```

```python
#Usando preprocessing
#Separando os dados
x=df.iloc[:,0:4].values
y=df.iloc[:,4].values
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
x[:,0] = label_encoder.fit_transform(x[:,0])
x[:,1] = label_encoder.fit_transform(x[:,1])
x[:,2] = label_encoder.fit_transform(x[:,2])
x[:,3] = label_encoder.fit_transform(x[:,3])
```

```python
x
```

```python
#Importar a arvore de decisao
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore_risco_credito = DecisionTreeClassifier(criterion='entropy')
#Treinando..
arvore_risco_credito.fit(x,y)
```

```python
arvore_risco_credito.feature_importances_ #Esta treinada
```

```python
arvore_risco_credito.classes_
```

```python
#Usando grafico para visualizar
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
previsores = ['historia', 'divida', 'garantias','renda']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(arvore_risco_credito, feature_names=previsores, class_names=arvore_risco_credito.classes_, filled=True)
```

```python
previsoes = arvore_risco_credito.predict([[0,0,1,2]])
previsoes
```
