### Clusterização dos dados

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('curriculos.csv')
df.head()
```

```python
# Excluindo colunas categóricas
df = df.drop(columns=['ANO-DOUTORADO','NUMERO-IDENTIFICADOR','CLASSE'], axis = 1)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

Artigos  = df['ARTIGO-PUBLICADO'].values
Eventos  = df['TRABALHO-EM-EVENTOS'].values
Projetos = df['PARTICIPACAO-EM-PROJETO'].values

X = np.array(list(zip(Artigos, Eventos, Projetos)))
plt.scatter(Artigos, Eventos, c='black', s=100)
```

```python
##plot in 3D space
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1])
```

```python
# U elbow method to find the ideal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

### Clusters e Centroide com redução de dimencionalidade PCA

```python
from sklearn.decomposition import PCA
```

```python
kmeans = KMeans(n_clusters=3, random_state=42).fit(df.values)
```

```python
kmeans_1 = pd.DataFrame(kmeans.labels_, columns=['cluster'])
```

```python
# Informaremos ao PCA em quantas dimenções queremos prjetar nossos dados
pca = PCA(n_components=2).fit(df)
```

Temos que tranformar nossos dados em no novo espaço e adicionar a tranformação em nosso conjunto de dados.

```python
# Tranformando nossos dados no novo espaço
pca_trans = pca.transform(df)

# Setando nome nas novas tranformações doda dos
pca_trans_df = pd.DataFrame(pca_trans, columns=['pca1', 'pca2'])

# Contatena nossos dados
kmeans_1 = pd.concat([kmeans_1, pca_trans_df], axis=1)
print(kmeans_1.head())
```

```python
pca1 = pca_trans_df['pca1']
pca2 = pca_trans_df['pca2']
```

```python
color = ['b','g','s']
markers = ['s', 'v']
```
