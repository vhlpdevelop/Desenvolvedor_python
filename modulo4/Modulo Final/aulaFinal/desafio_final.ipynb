```python
#imports
import pandas as pd
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
label_encoder = LabelEncoder()
df = pd.read_csv("census.csv")

```

```python
x=df.iloc[:,0:14].values
y=df.iloc[:,14].values
```

```python
x[:,1] = label_encoder.fit_transform(x[:,1])
x[:,3] = label_encoder.fit_transform(x[:,3])
x[:,5] = label_encoder.fit_transform(x[:,5])
x[:,6] = label_encoder.fit_transform(x[:,6])
x[:,7] = label_encoder.fit_transform(x[:,7])
x[:,8] = label_encoder.fit_transform(x[:,8])
x[:,9] = label_encoder.fit_transform(x[:,9])
x[:,13] = label_encoder.fit_transform(x[:,13])
x
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split

X_census_treinamento, X_census_teste, Y_census_treinamento, Y_census_teste = train_test_split(x, y, test_size=0.20, random_state = 0)

random_forest_census = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state = 0)

random_forest_census.fit(X_census_treinamento, Y_census_treinamento)
```

```python
previsoes = random_forest_census.predict(X_census_teste)
previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(Y_census_teste, previsoes)
```
